For the last case study, let's dive into more complex mechanics. Let's look at the plane wing as a beam, which is a more complicated model than a mass-spring system. For the beam it is possible to write the equation of oscillation in a matrix form, which is very similar to what you have already seen for the mass-spring system. 

$$M \mathbf{\ddot u} + k\mathbf{u}= 0$$

Run the code below to load the values of the $M$ and $k$ matrix. Feel free to print a few entries of the matrix to get an idea of what you are working with.

```python
data = loadmat('matrices_30/matrices_30.mat')

K = data['K']
M = data['M']

print(K.shape, M.shape)
```

For this exercise we will be calculating the diagonalization of A. 

To start, let us explore the symmetry of $M^{-1}$ and $M^{-1} * K$. What can we conclude about the symmetry of multiples of two symmetric matrices?

```python
M_inv = np.linalg.inv(M)
M_inv_K = M_inv @ K

is_symmetric = np.allclose(M_inv, M_inv.T)
print(is_symmetric)

is_symmetric_m_inv_k = np.allclose(M_inv_K, M_inv_K.T)
print(is_symmetric_m_inv_k)
```

Now, calculate $A$ in the code block below.

Hint: Remember that
 $$\begin{aligned} MM^{-1}\mathbf{\ddot u}+M^{-1}k\mathbf{u} &= \mathbf{0} \\ \mathbf{\ddot u}+A\mathbf{u} &= \mathbf{0} \end{aligned}$$

```python
A = np.matmul(np.linalg.inv(M), K)
```

$A$ has eigenvectors and eigenvalues; $A\mathbf{v_i} = \lambda_i\mathbf{v_i}$. We will use $\lambda_i$ = $\omega_i^2$ to match the notation used in physics. Now using $A$, calculate $P$ in the code block below.

```python
lambda_vals, P = np.linalg.eig(A)
```

Finally, using $A$  and $P$, calculate $D$ in the code block below.

```python
D = np.diag(lambda_vals)

# # alternatively:
# D = (np.linalg.inv(P) @ A ) @ P
# # to avoid very small floating point values:
# D_1 = np.round((np.linalg.inv(P) @ A ) @ P,decimals=1)
```

You can check your implementation by running the following assertion:

```python
diagonals = np.sort([D[i][i] for i in range(0,30)])
diagonals_ans = np.sort([data['D'][i][i] for i in range(0,30)])

np.testing.assert_allclose(diagonals, diagonals_ans, atol=0.1, err_msg="diagonals of D not correct")
```

Now, we can solve the decoupled system to calculate $u$ at a time $t$ and plot it. Run the code below to set your initial $u$ and $v$ vectors.

```python
rng = np.random.default_rng(42)
# randomly generate starting state
u_0 = rng.standard_normal(30) 
ud_0 = rng.standard_normal(30) 

# initialize time range to consider
t = np.linspace(0,10,50)
```

<!-- Now, calculate $u$ at $t=0.25$ for each point and use the function `plot_v` to plot your results.  -->
Now calculate the coefficients $c_1$ and $c_2$. Remember, $y=P^{-1}u$.

```python
eig_vals = np.asarray(np.sqrt(lambda_vals))

y_0 = np.linalg.inv(P) @ u_0
yd_0 = np.linalg.inv(P) @ ud_0

c_1 = y_0.copy()
c_2 = yd_0/eig_vals
```

Calculate the $y$ and $u$ vectors.

```python
c_1 = c_1.reshape(-1,1)
c_2 = c_2.reshape(-1,1)

theta = eig_vals[:, None] * t[None, :]

y_t = c_1 * np.cos(theta) + c_2 * np.sin(theta)
u_t = P @ y_t
```

Run the code below to visualize an element of $y$ at a certain time.

```python
time_val = 1

plt.plot(y_t[:,time_val],marker='o')
plt.title(f"y at time t={time_val}")
plt.grid(True)
plt.tight_layout()
plt.show()
```

Using the following functions, visualize the displacement of a certain point over time or the displacements of all ten points at a specific time. What do you notice?

```python
def plot_displacement_over_time(u_t,p=1):
    index_horiz = (p-1) * 3 
    index_vert = (p-1) * 3 + 1

    plt.subplot(2, 1, 1)
    plt.plot(u_t[index_horiz],marker='o')
    plt.title(f"Horizontal displacement of point {p} over time")
    plt.xlabel("Time")
    plt.ylabel("Displacement")
    plt.grid(True)
    plt.tight_layout()
    plt.subplot(2, 1, 2)

    plt.plot(u_t[index_vert],marker='o')
    plt.title(f"Vertical displacement of point {p} over time")
    plt.xlabel("Time")
    plt.ylabel("Displacement")
    plt.grid(True)

    plt.suptitle(f"Displacement of Point {p} over Time")
    plt.tight_layout()
    plt.show()


def plot_points_at_time(u_t,t=1):
    indices_horiz = np.arange(0,len(u_t),3)
    indices_vert = np.arange(1,len(u_t),3)

    plt.subplot(2, 1, 1)
    plt.plot(u_t[indices_horiz,t],marker='o')
    plt.title(f"Horizontal displacement of points 1-10 at t={t}")
    plt.xlabel("Time")
    plt.ylabel("Displacement")
    plt.grid(True)
    plt.tight_layout()
    plt.subplot(2, 1, 2)

    plt.plot(u_t[indices_vert,t],marker='o')
    plt.title(f"Vertical displacement of points 1-10 at t={t}")
    plt.xlabel("Time")
    plt.ylabel("Displacement")
    plt.grid(True)

    plt.suptitle(f"Displacement of Points 1-10 at t={t}")
    plt.tight_layout()
    plt.show()

plot_displacement_over_time(u_t)
plot_points_at_time(u_t)
```
